# **Reading csv file**

```python
df = pd.read_csv('/content/drive/MyDrive/obdusingcsv/objects (1).csv')
```

# **Dropping all the irrelevent columns with null values and unusefull data**

```python
df.drop(['subclass' , 'convex' , 'headtail' , 'orient' , 'relevance' , 'mark_tool'] , axis = 1 , inplace = True)
```

# **Splitting the bbox column and concatinating into the dataframe**

```python
df = pd.concat([df , df['bbox'].str.split(expand = True)] , axis = 1)
```

# **renaming the bbox columnd after split and concatination**

```python
df.rename(columns = {0 : 'xmin' , 1 : 'xmax' , 2 : 'ymin' , 3 : 'ymax'} , inplace = True)
```

# **Splitting the center column and concatinating into the dataframe**

```python
df = pd.concat([df , df['center'].str.split(expand = True)] , axis = 1)
```

# **renaming the center columnd after split and concatination**

```python
df.rename(columns = {0 : 'center_width' , 1 : 'center_height'} , inplace = True)
```

```python
df.head()
```

# **Dropping old bbox and center columns**

```python
df.drop(['center' , 'bbox'] , axis = 1 , inplace =  True)
```

# **Using bounding box formulae to conver pixel size to bounding box format of 0 and 1**

```python
df1['xmin'] = df1['xmin'] / df1['center_width']
df1['xmax'] = df1['xmin'] / df1['center_height']
df1['ymin'] = df1['xmin'] / df1['center_width']
df1['ymax'] = df1['xmin'] / df1['center_height']
```

# **writing loop to iterate the rows in the id coulumn and each file with seperate file name of each row and saving calculated bounding box information with class in a specified location**

```python
for index , row in df1.iterrows() :
  file_name = "/content/drive/MyDrive/obdusingcsv/leopards labels/"+row['id'].split(".")[0]+".txt"
  with open (file_name , 'w') as file :

    mystr = '0 ' + str(row['xmin']) + ' ' + str(row['xmax']) + ' ' + str(row['ymin']) + ' ' + str(row['ymax'])
    file.write(mystr)
    print(row['id'])
```
